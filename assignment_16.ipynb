```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# reading the data file

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

# scatter graph plot using dot

```python
sns.relplot(x='parental level of education',y='math score',data=df,hue='gender',style='lunch')
```

# scatter graph plot using line

```python
sns.relplot(x='parental level of education',y='math score',data=df,hue='gender',style='lunch',kind='line')
```

```python
sns.catplot(x='parental level of education',y='gender',data=df)
```

# Bar Graph

```python
sns.catplot(x='parental level of education',y='math score',data=df,kind='bar')
```

# scatter graph

```python
sns.scatterplot(x='gender',y='math score',data=df)
```

# bar graph

```python
sns.barplot(x='parental level of education',y='math score',data = df,hue='gender')
plt.title('bar graph')
plt.show()
```

# creating list 

```python
level = list(df['parental level of education'].value_counts().keys())
values = list(df['parental level of education'].value_counts().values)
level
values
```

# Pie Chart

```python
plt.pie(values,labels=level,wedgeprops={'ec':'yellow'},autopct='%1.2f%%',shadow=True)
plt.title('pie chart')
plt.show()
```
